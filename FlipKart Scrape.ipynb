```python
import requests
from bs4 import BeautifulSoup
```

```python
URL = "https://www.flipkart.com/search?q=mobiles&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off"
```

```python
response = requests.get(URL)
```

```python
response.status_code
```

```python
content = response.text
```

```python
content[:1000]
```

```python
with open('webpage.html', 'w', encoding= 'utf-8') as f:
    f.write(content)
```

 ## Beautiful Soup

```python
soup = BeautifulSoup(content, 'html.parser')
```

```python
soup.prettify()
```

##### Mobile Names

```python
p_tags = soup.find_all('div')
len(p_tags)
```

```python
mobile_names = soup.find_all('div', {'class': '_4rR01T'})
len(mobile_names)
```

```python
for names in mobile_names:
    print(names.text)
```

```python
mobile_names_list = []

for names in mobile_names:
    mobile_names_list .append(names.text)
```

```python
mobile_names_list 
```

###### Mobile Prices

```python
mobile_prices = soup.find_all('div', {'class': '_1_WHN1'})
len(mobile_prices)
```

```python
for prices in mobile_prices:
    print(prices.text[1:])
```

```python
mobile_price_list = []

for prices in mobile_prices:
    mobile_price_list .append(prices.text[1:])
```

```python
mobile_price_list
```

##### Mobile Ratings

```python
mobile_ratings = soup.find_all('div', {'class': '_3LWZlK'})
len(mobile_ratings)
```

```python
for ratings in  mobile_ratings[:24]:
    print(ratings.text)
```

```python
mobile_ratings_list = []


for ratings in  mobile_ratings[:24]:
    mobile_ratings_list.append(ratings.text)
```

```python
mobile_ratings_list
```

##### Mobile Company

```python
for names in mobile_names_list:
    print(names.split(" ")[0])
```

```python
mobile_comapany_list = []


for company_names in mobile_names_list:
    mobile_comapany_list.append(company_names.split(" ")[0])
```

```python
mobile_comapany_list
```

##### Mobile Features

```python
mobile_features= soup.find_all('ul', {'class': '_1xgFaf'})
len(mobile_features)
```

```python
mobile_feature_list=[]


for feature_list in mobile_features:
    mobile_feature_list.append(feature_list.text)
```

```python
 mobile_feature_list_updated =[]
    
    
for feature in mobile_feature_list:
    mobile_feature_list_updated.append(feature.split("|"))
```

```python
 mobile_feature_list_updated 
```

##### RAM Column

```python

ram_list=[]

for ram in mobile_feature_list_updated:
    ram_list.append(ram[0])
    
ram_list
```

```python
rom_list=[]

for rom in mobile_feature_list_updated:
    rom_list.append(rom[1])
```

```python
rom_list
```

```python
rom_list_updated=[]

for i in rom_list:
        rom_list_updated.append(i[:7])
rom_list_updated 
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(list(zip(mobile_comapany_list,mobile_names_list ,ram_list,rom_list_updated,mobile_price_list,mobile_ratings_list)), columns = ["Company", "Model","RAM","ROM","Price(rupees)","Ratings"]) 
```

```python
df
```

```python
df.to_csv("Flipkart_mobiles.csv", index=None)
```
